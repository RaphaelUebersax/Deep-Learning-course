### Plotting learning rates
This file is used to, given the results in lr_results.pt, plot them to choose the best learning rate for each model.
It stores the resulting plot in stored_results/lr_plot.png

```python
import torch
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
def seaborn_lr():
    font_title = 17
    font = 15
    results_to_plot = torch.load('stored_results/lr_results.pt')
    print(results_to_plot.shape)

    fig = plt.figure(figsize=(7,7))
    #fig.suptitle('Learning rate comparison',fontsize=20)
    sns.set_style("darkgrid",{"axes.edgecolor": ".1"})
    gs = plt.GridSpec(nrows=2, 
                       ncols=2, 
                       figure=fig, 
                       width_ratios= [1, 1],
                       height_ratios=[1, 1],
                       wspace=0.2,
                       hspace=0.3
                    )


    ax1 = fig.add_subplot(gs[0,0])
    ax1.set_title("Net", fontsize=font_title)
    l1 = sns.lineplot(data=results_to_plot[0,0,:])
    l2 = sns.lineplot(data=results_to_plot[0,1,:])
    l3 = sns.lineplot(data=results_to_plot[0,2,:])
    l4 = sns.lineplot(data=results_to_plot[0,3,:])
    l5 = sns.lineplot(data=results_to_plot[0,4,:])
    l6 = sns.lineplot(data=results_to_plot[0,5,:])
    ax1.lines[2].set_linestyle("--")
    ax1.set_yscale('log')
    ax1.set_ylabel("Loss", fontsize=font)

    ax2 = fig.add_subplot(gs[0,1])
    ax2.set_title("LeNet5", fontsize=font_title)
    sns.lineplot(data=results_to_plot[1,0,:])
    sns.lineplot(data=results_to_plot[1,1,:])
    sns.lineplot(data=results_to_plot[1,2,:])
    sns.lineplot(data=results_to_plot[1,3,:])
    sns.lineplot(data=results_to_plot[1,4,:])
    sns.lineplot(data=results_to_plot[1,5,:])
    ax2.lines[3].set_linestyle("--")
    ax2.set_yscale('log')

    ax3 = fig.add_subplot(gs[1,0])
    ax3.set_title("LeNet5_FullyConv", fontsize=font_title)
    sns.lineplot(data=results_to_plot[2,0,:])
    sns.lineplot(data=results_to_plot[2,1,:])
    sns.lineplot(data=results_to_plot[2,2,:])
    sns.lineplot(data=results_to_plot[2,3,:])
    sns.lineplot(data=results_to_plot[2,4,:])
    sns.lineplot(data=results_to_plot[2,5,:])
    ax3.lines[1].set_linestyle("--")
    ax3.set_yscale('log')
    ax3.set_xlabel("Epoch num.", fontsize=font)
    ax3.set_ylabel("Loss", fontsize=font)


    ax4 = fig.add_subplot(gs[1,1])
    ax4.set_title("ResNet", fontsize=font_title)
    sns.lineplot(data=results_to_plot[3,0,:])
    sns.lineplot(data=results_to_plot[3,1,:])
    sns.lineplot(data=results_to_plot[3,2,:])
    sns.lineplot(data=results_to_plot[3,3,:])
    sns.lineplot(data=results_to_plot[3,4,:])
    sns.lineplot(data=results_to_plot[3,5,:])
    ax4.lines[3].set_linestyle("--")
    ax4.set_yscale('log')
    ax4.set_xlabel("Epoch num.", fontsize=font)
    

    lgd = ax3.legend(labels=['3e-1', '5e-2', '3e-2', '5e-3', '3e-3', '5e-4'],loc='upper center', bbox_to_anchor=(0.5, -0.2), fancybox=True, ncol=3)

    plt.savefig('stored_results/lr_plot.png', dpi=300, bbox_extra_artists=(lgd,), bbox_inches='tight')
    plt.show()
```

```python
seaborn_lr()
```
